```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load audio
audio, sr = librosa.load("on_0.wav", sr=16000)

# Extract MFCCs with 13 and 40 coefficients
mfcc_13 = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=13)
mfcc_40 = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=40)

# Plot
plt.figure(figsize=(14, 8))

# First: 13 MFCCs
plt.subplot(2, 1, 1)
librosa.display.specshow(mfcc_13, x_axis='time', sr=sr)
plt.colorbar()
plt.title("MFCCs (n_mfcc=13)")

# Second: 40 MFCCs
plt.subplot(2, 1, 2)
librosa.display.specshow(mfcc_40, x_axis='time', sr=sr)
plt.colorbar()
plt.title("MFCCs (n_mfcc=40)")

plt.tight_layout()
plt.show()
```
